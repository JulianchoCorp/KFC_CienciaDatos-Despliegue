```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.ticker as ticker
```

```python
cwd = os.getcwd()
raw_dir = cwd + "/../Datos/raw/"
processed_dir = cwd + "/../Datos/processed/"
graphs_dir = cwd + "/../graphs/"
```

```python
import dash

print(dash.__version__)
```

```python
df_unificado = pd.read_csv(processed_dir + 'datos_limpios_unidos.csv', parse_dates=['fecha_trans'])
```

```python

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(10,5))
# Calcular la cantidad de transacciones por restaurante
transacciones_por_restaurante = df_unificado['nom_rest'].value_counts(ascending=True)

sns.color_palette("muted")
# Crear el grfico de barras utilizando Seaborn
sns.barplot( x=transacciones_por_restaurante.values,y=transacciones_por_restaurante.index, color='c')

ax.set( ylabel='Cantidad de transacciones',
       xlabel='Restaurante')
# Personalizar el grfico
plt.xticks(rotation='vertical')
plt.title('Cantidad de transacciones por restaurante')
sns.despine(left=True, bottom=True)
# Mostrar el grfico
sns.despine(left=True, bottom=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el estilo del grfico
sns.set_theme(style="ticks")

# Agrupar por canal y restaurante y calcular el total de transacciones
total_transacciones = df_unificado.groupby(['canal', 'nom_rest'])['cod_trans'].count().reset_index()

# Crear el grfico de barras apiladas
g = sns.catplot(
    data=total_transacciones, kind="bar",
    y="nom_rest", x="cod_trans", hue="canal",
    palette="dark:red_r", height=8
)


# Personalizar el grfico
g.despine(left=True)
g.set_axis_labels("Total de Transacciones - Miles", "Restaurante")
g.legend.set_title("Canal")

# Obtener el objeto de los ejes del grfico
ax = g.ax

# Formatear el eje x en miles
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1000:.0f}'))

# Resaltar las barras ms grandes haciendo que sean ms gruesas
max_value = total_transacciones['cod_trans'].max()
for bar in ax.patches:
    height = bar.get_height()
    if height == max_value:
        bar.set_linewidth(2)  # Ancho de lnea de las barras ms grandes
    else:
        bar.set_linewidth(1)  # Ancho de lnea de las dems barras

# Mostrar el grfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el estilo del grfico
sns.set_theme(style="ticks")

# Agrupar por canal y restaurante y calcular el total de transacciones
total_transacciones = df_unificado.groupby(['canal', 'nom_rest'])['cod_trans'].count().reset_index()

# Ordenar las transacciones de forma descendente
total_transacciones = total_transacciones.sort_values('cod_trans', ascending=False)

# Crear el grfico de barras apiladas
g = sns.catplot(
    data=total_transacciones, kind="bar",
    y="nom_rest", x="cod_trans", hue="canal",
    palette=["lightgrey" if x not in total_transacciones.head(5)['nom_rest'].values else "red" for x in total_transacciones['nom_rest']],
    height=10
)
plt.xticks(rotation='vertical')

# Personalizar el grfico
g.despine(left=True)
g.set_axis_labels("Transacciones - Miles"Restaurante")
g.legend.set_title("Canal")

# Quitar el eje x

ax = g.axes[0][0]
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1000:.0f}'))

# Mostrar el grfico
plt.show()
```

```python
# Crear el grfico de dispersin utilizando Seaborn
sns.scatterplot(x=df_unificado['longitud'], y=df_unificado['latitud'])

# Personalizar el grfico
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Distribucin geogrfica de las transacciones')

# Mostrar el grfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la suma de las comisiones por canal
comisiones_por_canal = df_unificado.groupby('canal')['valor_trans'].count()

# Crear el grfico de pastel utilizando Seaborn
plt.pie(comisiones_por_canal.values, labels=comisiones_por_canal.index, autopct='%1.1f%%')

# Personalizar el grfico
plt.title('Proporcin de comisiones por canal')

# Mostrar el grfico
plt.show()
```
